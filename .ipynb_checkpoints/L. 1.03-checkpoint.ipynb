```python
# correcting data types
dataset['median_home_val'] =  pd.to_numeric(dataset['median_home_val'], errors='coerce')
dataset['ic5'] =  pd.to_numeric(dataset['ic5'], errors='coerce')
dataset.head(20)

dataset._get_numeric_data() # to check if 'median_home_val' and 'ic5' are now listed as numeric data
```

```python
# Removing duplicates
print(dataset.shape)
data2 = dataset.copy()
data2 = data2.drop_duplicates()  # nb keep argument
print(data2.shape)
# #if we want to remove duplicates based on some specific columns
#data2 = data2.drop_duplicates(subset=['state','gender', 'ic2', 'ic3'])
data2 = data2.drop_duplicates(subset=['state','gender', 'ic2', 'ic3'],keep='last')
## #data2
print(data2.shape)

```
